## MTA data

```python
import pandas as pd
```

```python
#combine daily data
def get_mta_data(week_nums):
    filepath = "mta_{}_daily.csv"
    dfs = []
    for year in years:
        file = filepath.format(year)
        dfs.append(pd.read_csv(file, sep='\t'))
    return pd.concat(dfs)
        
years=[2015,2016,2017,2018,2019,2020,2021,2022]
mta_df = get_mta_data(years)
```

```python
mta_df[mta_df.isna().any(axis=1)]
```

```python
mta_df
```

```python
mta_df.drop_duplicates(subset=["DATE"], inplace=True)
```

```python
mta_df
```

```python
mta_df["MONTH"] = pd.to_datetime(mta_df['DATE']).dt.strftime('%Y-%m')
```

```python
total_monthly=mta_df.groupby(["MONTH"]).DAILY_ENTRIES.sum().reset_index(name ='MONTHLY_ENTRIES')
```

```python
total_monthly.head()
```

```python
total_monthly.tail()
```

```python
#remove 2014 and 2022 data
total_monthly=total_monthly[~total_monthly.MONTH.isin(['2014-12','2022-01'])]
```

```python
#write monthly MTA data to .csv
total_monthly.to_csv('mta_total_monthly.csv', sep='\t', index=True)
```

## Uber data

```python
import numpy as np
```

```python
#https://data.cityofnewyork.us/Transportation/FHV-Base-Aggregate-Report/2v9c-2k7f
uberlyft_data=pd.read_csv('UberLyftData.csv')
```

```python
uberlyft_data=uberlyft_data[uberlyft_data['Base Name'].isin(['UBER','LYFT'])]
```

```python
uberlyft_data.head()
```

```python
uberlyft_data.dtypes
```

```python
#remove comma in number
uberlyft_data['Total Dispatched Trips']=uberlyft_data['Total Dispatched Trips'].str.replace(',', '').astype(np.int64)
```

```python
uberlyft_data['MONTH'] = uberlyft_data.Year.astype(str) + '-' + uberlyft_data.Month.astype(str)
```

```python
#convert date to yyyy-mm format
uberlyft_data["MONTH"] = pd.to_datetime(uberlyft_data['MONTH']).dt.strftime('%Y-%m')
```

```python
uberlyft_data=uberlyft_data.sort_values(["MONTH"], ascending=True)
```

```python
uberlyft_data
```

### Plot Monthly MTA & Uber/Lyft data

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
total_monthly['rolling_monthly_entries']=total_monthly.MONTHLY_ENTRIES.rolling(6).mean()
```

```python
total_monthly
```

```python
#Plotting NYC covid overlayed with MTA data
#plt.figure( figsize = (10, 10))

#https://learndataanalysis.org/source-code-format-axis-label-to-thousands-and-millions-suffix-matplotlib-tutorial/
#https://stackoverflow.com/questions/61330427/set-y-axis-in-millions
#Format Axis Label To Thousands and Millions Suffix
def millions_formatter(x, pos):
    return f'{x / 1000000}M'


fig, ax = plt.subplots(figsize=(12,8))


xticks=total_monthly.MONTH[::5]
xlabels=total_monthly.MONTH[::5]

ax.set_facecolor('white')
sns.set_style("whitegrid")
sns.set_context("poster")
sns.despine()

ax=sns.lineplot(data=total_monthly,
                x = 'MONTH',
                y = 'MONTHLY_ENTRIES',
                label = 'MTA Monthly Entries',
                legend=False,
                color='darkturquoise'
               )

sns.lineplot(data=total_monthly,
             x = 'MONTH',
             y = 'rolling_monthly_entries',
             label = 'MTA Monthly Rolling',
             ax=ax,
             color="goldenrod",
             linestyle='--',
             legend=False
             )

ax2 = ax.twinx()
                
sns.lineplot(data=uberlyft_data,
             x = 'MONTH',
             y = 'Total Dispatched Trips',
             label = 'Uber/Lyft Monthly Trips',
             ax=ax2,
             #color="mediumorchid",
             hue="Base Name",
             palette={'LYFT':'magenta','UBER':'black'},
             legend=False
             )

#ax.yaxis.grid(False)
#ax2.yaxis.grid(False)
#ax.xaxis.grid(False)
#ax2.xaxis.grid(False)

#xticks=NYC_COVID_Data.date[::90] #only show tick for every 90 days
#xlabels=NYC_COVID_Data.abbrev_date[::90] #only show tick for every 90 days

#ax.set_xticklabels(labels=xlabels, rotation=30)

#plt.xlabel("aergaerg"
           #, fontsize = 20
          #)

#plt.ylabel("Daily Cases"
           #, fontsize = 20
          #)

ax.set_ylabel('Monthly MTA Entries', fontsize = 15)
ax2.set_ylabel('Monthly Uber/Lyft Entries', fontsize = 15)
ax.set_xlabel('Date', fontsize = 15)


#plt.tight_layout()
#plt.legend(fontsize = 20)

plt.title("2019-2022 NYC Ridershare MTA Data"
          #, fontsize = 30
         )

fig.legend(labels=["MTA",'MTA 6 Month Rolling',"Uber",'Lyft'],fontsize = 15)

ax2.set_ylim(0,80000000)


ax.xaxis.set_ticks(xticks)
ax.xaxis.set_ticklabels(xlabels, rotation=45, fontsize = 15)

ax.set_yticklabels(ax.get_yticks(), size = 13)
ax2.set_yticklabels(ax2.get_yticks(), size = 13)
#ax.set_xticklabels(ax2.get_xticklabels(), rotation=45)

ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
ax2.yaxis.set_major_formatter(FuncFormatter(millions_formatter))

fig.tight_layout()
sns.set(font_scale = 1)

fig.figure.savefig("Uber_MTA.png", dpi=1200)
```
