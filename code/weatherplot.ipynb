## MTA data

```python
import numpy as np
```

```python
mta_daily=pd.read_csv('../data/mta_total_daily.csv', sep='\t', index_col=0)
```

```python
mta_daily.head()
```

```python
mta_daily['DATE'] = pd.to_datetime(mta_daily['DATE'])
```

```python
mta_daily['day_of_week'] = mta_daily['DATE'].dt.day_name()
mta_daily.head()
```

```python
mta_daily['isweekend'] = np.where(mta_daily['day_of_week'].isin(['Sunday','Saturday']), 'Weekend', 'Weekday')
```

```python
mta_daily.head(14)
```

```python
mta_daily['issummer'] = np.where(mta_daily['DATE'].dt.month.isin([7,8]), 'Summer', 'Not_Summer')
```

```python
mta_daily.head(200)
```

## Merging and plotting data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#convert DATE column to datetime type otherwise get ValueError when merge datetime with object type
weather_data_mean['DATE'] = pd.to_datetime(weather_data_mean['DATE'])
```

```python
df = pd.merge(mta_daily, weather_data_mean, on='DATE', how='inner')
```

```python
df
```

```python
#precipitation vs mta
ax = sns.regplot(x="PRCP", y="DAILY_ENTRIES", data=df)
```

```python
df_zero=df.fillna(0)
```

```python
#precipitation vs mta
#nan replaced with zeros
ax = sns.regplot(x="PRCP", y="DAILY_ENTRIES", data=df_zero)
```

```python
#get list of independent variables
mylist=df_zero.columns.tolist()[9:]
len(mylist)
```

```python
fig, ax = plt.subplots(nrows=15,ncols=2,figsize=(10,100))

for i,var in enumerate(mylist[:15]):
    #state_cpm = state + "_" + "cpm"
    #state_spend = state + "_" + "spend"

    sns.regplot(data=df_zero, x=var, y="DAILY_ENTRIES",
                ax=ax[i,0]
               )
    
for i,var in enumerate(mylist[15:]):
    #state_cpm = state + "_" + "cpm"
    #state_spend = state + "_" + "spend"

    sns.regplot(data=df_zero, x=var, y="DAILY_ENTRIES",
                ax=ax[i,1]
               )

    
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
#final plots
#https://learndataanalysis.org/source-code-format-axis-label-to-thousands-and-millions-suffix-matplotlib-tutorial/
#https://stackoverflow.com/questions/61330427/set-y-axis-in-millions
#Format Axis Label To Thousands and Millions Suffix
def millions_formatter(x, pos):
    return f'{x / 1000000}M'


fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(15,5))

fig.suptitle('Weather vs MTA Ridership (2017-2019)')


sns.regplot(data=df_zero, x='PRCP', y="DAILY_ENTRIES",
            ax=ax[0]
            )
ax[0].set_xlabel( "Precipitation" , size = 12 )
ax[0].set_ylabel( "Daily MTA Entries" , size = 12 )
#ax[0].set_title( "Bill vs Tips" , size = 24 )


sns.regplot(data=df_zero, x='SNWD', y="DAILY_ENTRIES",
            ax=ax[1]
            )
ax[1].set_xlabel( "Snow Depth" , size = 12 )
ax[1].set_ylabel( "" , size = 12 )
#ax[1].set_title( "Bill vs Tips" , size = 24 )


sns.regplot(data=df_zero, x='TAVG', y="DAILY_ENTRIES",
            ax=ax[2]
            )  
ax[2].set_xlabel( "Average Temperature" , size = 12 )
ax[2].set_ylabel( "" , size = 12 )
#ax[2].set_title( "Bill vs Tips" , size = 24 )

ax[0].yaxis.set_major_formatter(FuncFormatter(millions_formatter))
ax[1].yaxis.set_major_formatter(FuncFormatter(millions_formatter))
ax[2].yaxis.set_major_formatter(FuncFormatter(millions_formatter))

sns.set(font_scale = 2)
fig.figure.savefig("Weather_MTA.png", dpi=1200,transparent=True)
```

```python
#zero in on temperature plot
def millions_formatter(x, pos):
    return f'{x / 1000000}M'

fig = plt.figure(figsize=(10,8))

ax=sns.scatterplot(data=df_zero, x='TAVG', y="DAILY_ENTRIES",
           hue='issummer'
            )  

ax.set_xlabel( "Average Temperature" )
ax.set_ylabel( "Daily MTA Entries" )
plt.legend(
            #labels=["1","2","3","4","5"], 
           loc = 2, 
           bbox_to_anchor = (1,1)
          )

ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))

sns.set(font_scale = 2)
fig.figure.savefig("Temperature_MTA.png", dpi=1200, bbox_inches="tight",transparent=True)
```
